```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
import plotly.graph_objs as go
import plotly.figure_factory as ff
from plotly import tools
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

## Variables distribiutions and correlations 

```python
df[df.columns.to_list()[1:-2]].boxplot(rot=45,figsize=(20,20))
```

```python
for index,i in enumerate(df.columns.to_list()[1:]):
    plt.figure()
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(16,5))
    sns.violinplot(df[i][df['Class'] == 0],ax=ax1).set_title('No fraud')
    sns.violinplot(df[i][df['Class'] == 1],ax=ax2).set_title('Fraud')
    plt.show()
```

```python
sns.countplot(df['Class']).set_title('Class counts')
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(16,8))

sns.boxplot(ax=ax1,x='Class',y='Amount',data=df,hue='Class',showfliers=True)
sns.boxplot(ax=ax2,x='Class',y='Amount',data=df,hue='Class',showfliers=False)
plt.show()
```

```python
class_0 = df.query('Class == 0')['Time']
class_1 = df.query('Class == 1')['Time']

hist_data = [class_0,class_1]
names = ['Not fraud','Fraud']

fig = ff.create_distplot(hist_data,names,show_hist=False, show_rug=False)
fig['layout'].update(title='Credit Card Transactions Time Density Plot', xaxis=dict(title='Time [s]'))
iplot(fig, filename='dist_only')
```

```python
plt.subplots(1,1,figsize=(20,20))
sns.heatmap(df.corr(),linewidths=.1,cmap="RdYlBu")
```
